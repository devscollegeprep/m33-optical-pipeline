```python
single_exposure_time = 120.0
```

* Stores the duration of one telescope shot
* A float (number with a decimal point)

The math: the total integrated charge Q accumulated on one pixel sensor over an exposure duration $t_{\text{exp}}$ is defined by the definite integral of the photon arrival rate R(t)

$$Q = \int_{0}^{t_{\text{exp}}} R(t) \, dt$$

* Because the rig holds continuous, steady exposure, $t_{\text{exp}} = 120.0\,\text{s}$ is the upper limit of integration

$$T_{\text{total}} = \sum_{i=1}^{N} t_{\text{exp}}$$
